```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('data.csv')
df.head()

# Updating the header
df.columns=["date","RK"]
df.head()
df.describe()
df.set_index('date',inplace=True)

from pylab import rcParams
rcParams['figure.figsize'] = 15, 7
df.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_result=adfuller(df['RK'])
```

```python
def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
```

```python
df['Sales First Difference'] = df['RK'] - df['RK'].shift(1)
df['Seasonal First Difference']=df['RK']-df['RK'].shift(12)
df.head()
```

```python
adfuller_test(df['Seasonal First Difference'].dropna())
```

```python
df['Seasonal First Difference'].plot()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df['RK'])
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(df['RK'],order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
```

```python
df['forecast']=model_fit.predict(start=10,end=22,dynamic=True)
df[['RK','forecast']].plot(figsize=(12,8))
```

```python
import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(df['RK'],order=(1, 1, 1),seasonal_order=(1,1,1,5))
results=model.fit()
```

```python
df['forecast']=results.predict(start=10,end=22,dynamic=True)
df[['RK','forecast']].plot(figsize=(12,8))
```
